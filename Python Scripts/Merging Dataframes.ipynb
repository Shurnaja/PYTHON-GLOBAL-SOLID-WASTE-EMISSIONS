# Changing value names

```python
#Changing names of the countries in WMG

name_mapping_1 = {'Dem. Rep. Congo':'Democratic Republic of the Congo', 'Republic of Congo': 'Congo', 'United Kingdom': 'United Kingdom of Great Britain and Northern Ireland'}
```

```python
df_wmg['Country'] = df_wmg['Country'].replace(name_mapping_1)
```

```python
#Exporting the WMG dataframe
df_wmg.to_csv(os.path.join(path, 'Prepared Data', 'WMG_wrangeled_2.csv'))
```

```python
df_wmg['Country'] = df_wmg['Country'].str.replace("'", "")
```

```python
#Renaming values in the dataframe 'Solid Waste', column 'Country'

name_mapping_2 = {'Cote DIvoire':'CÃ´te dIvoire'}
```

# Merging


```python
df_all_merged = pd.merge(df_waste, df_pop, on=['Country', 'Year'], how = 'left')
```

```python
df_all_merged_1 = pd.merge(df_all_merged, df_gdp, on=['Country', 'Year'], how = 'left')
```

```python
df_all_merged_2 = pd.merge(df_all_merged_1, df_wmg, on=['Country', 'Year'], how = 'left')
```
